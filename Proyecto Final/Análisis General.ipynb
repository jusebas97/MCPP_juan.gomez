Importamos los archivos de los repositorios de cada universidad

```python
df_1 = pd.read_pickle('repo_javeriana.pkl')
```

```python
df_2 = pd.read_pickle('repo_andes.pkl')
```

```python
df_3 = pd.read_pickle('repo_UR.pkl')
```

Unimos todos los repositorios dentro de un solo DataFrame, para obtener las 3506 tesis

```python
df = pd.concat([df_1, df_2, df_3, df_4], axis=0)
```

```python
df.info()
```

Calculamos las tesis total por año y graficamos

```python
txa = df.groupby(['Fecha']).size()
```

```python
txa.plot(kind='bar', stacked=True, color = 'b')
plt.grid()
plt.xlabel('Años')
plt.ylabel('Cantidad de tesis')
plt.title('Total de tesis por año - 4 Universidades')
plt.rc('xtick', labelsize=10) 
plt.rc('ytick', labelsize=10) 

plt.savefig('txa_global.png')
```

Calculamos la frecuencia de los Bigramas en el total de las tesis

```python
bigrams = df['bigrams'].tolist()
bigrams = list(chain(*bigrams))
bigrams = [(x.lower(), y.lower()) for x,y in bigrams]
bigram_counts = list(Counter(bigrams).most_common(15))
bigram_counts = tuple(bigram_counts)
```

Graficamos

```python
df_bigrams = pd.DataFrame(bigram_counts, columns=['Bigrama', 'Frecuencia'])
df_bigrams.plot(kind='bar', x = 'Bigrama', stacked=True, color='b')
plt.grid()
plt.xlabel('Bigramas/Temas')
plt.ylabel('Cantidad de tesis')
plt.title('Total de frecuencia de temas - 4 universidades')
plt.rc('xtick', labelsize=10) 
plt.savefig('Freq_bigramas_total.png')
```
