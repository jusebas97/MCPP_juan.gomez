## 6. [2]
Escriba un loop que imprima todos los números entre 6 y 30 que no son divisibles por 2, 3 o 5.

```python
for i in range (6,30):
    if (i % 2 != 0) :
        print (f"{i} no es divisible por 2")
    if (i % 3 != 0) :
        print (f"{i} no es divisible por 3")
    if (i % 5 != 0) :
        print (f"{i} no es divisible por 5")
```
